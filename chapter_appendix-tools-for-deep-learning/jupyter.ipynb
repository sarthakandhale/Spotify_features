## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

````python
# ============================================================
# üìò Assignment No. 1: Feature Engineering on Spotify Dataset
# ============================================================

# Import Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.feature_selection import VarianceThreshold
import matplotlib.pyplot as plt
import seaborn as sns

# ------------------------------------------------------------
# 1Ô∏è‚É£ Load and Explore Dataset
# ------------------------------------------------------------
# ‚ö†Ô∏è Make sure 'Features.csv' is in the same folder as this notebook.
df = pd.read_csv("/content/Features.csv")

print("‚úÖ Dataset Loaded Successfully!")
print("\nüìä First 5 Rows of the Dataset:\n")
print(df.head())

print("\nDataset Info:\n")
print(df.info())

print("\nMissing Values per Column:\n")
print(df.isnull().sum())

# ------------------------------------------------------------
# 2Ô∏è‚É£ Handle Missing Values
# ------------------------------------------------------------
# Replace numeric missing values with column mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# For non-numeric columns (like 'name'), fill with most frequent value
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\n‚úÖ Missing Values Handled Successfully!\n")

# ------------------------------------------------------------
# 3Ô∏è‚É£ Feature Encoding
# ------------------------------------------------------------
# Encode the 'name' column using its length (feature engineering idea)
df['name_length'] = df['name'].apply(lambda x: len(str(x)))

# Drop 'id' since it‚Äôs just an identifier
df.drop(columns=['id'], inplace=True)

print("‚úÖ Feature Encoding Completed (Added 'name_length')")

# ------------------------------------------------------------
# 4Ô∏è‚É£ Feature Scaling
# ------------------------------------------------------------
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[numeric_cols])

df_scaled = pd.DataFrame(scaled_data, columns=numeric_cols)
print("\n‚úÖ Feature Scaling Completed!\n")

# ------------------------------------------------------------
# 5Ô∏è‚É£ Feature Extraction (PCA)
# ------------------------------------------------------------
pca = PCA(n_components=0.95)  # Keep 95% of variance
pca_data = pca.fit_transform(df_scaled)

print("‚úÖ PCA Completed!")
print(f"Number of Components Retained: {pca.n_components_}")

# Optional: visualize explained variance
plt.figure(figsize=(6,4))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.title('PCA - Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# 6Ô∏è‚É£ Feature Selection (Correlation + Variance Threshold)
# ------------------------------------------------------------
# Check correlation between features
plt.figure(figsize=(10,8))
sns.heatmap(df_scaled.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
plt.show()

# Apply variance threshold
selector = VarianceThreshold(threshold=0.01)
selected_data = selector.fit_transform(df_scaled)

print("\n‚úÖ Feature Selection Completed!")
print(f"Number of Features Remaining: {selected_data.shape[1]}")

# ------------------------------------------------------------
# 7Ô∏è‚É£ Summary of Transformations
# ------------------------------------------------------------
print("\nüéØ Summary of Feature Engineering Steps:")
print("‚Ä¢ Missing values handled with mean/mode imputation.")
print("‚Ä¢ Added 'name_length' as a new feature.")
print("‚Ä¢ Standardized all numeric columns.")
print("‚Ä¢ Applied PCA for dimensionality reduction.")
print("‚Ä¢ Removed low-variance and redundant features.")
print("\nTransformed Dataset is ready for model building!")

# ------------------------------------------------------------
# 8Ô∏è‚É£ Save the Final Clean Dataset
# ------------------------------------------------------------
final_df = pd.DataFrame(selected_data)
final_df.to_csv("Spotify_Features_Transformed.csv", index=False)
print("\nüíæ Transformed dataset saved as 'Spotify_Features_Transformed.csv'")
````

## Prepare data for modeling

### Subtask:
Separate the features (X) and the target variable (y). Split the data into training and testing sets.


**Reasoning**:
Based on the previous steps and the nature of the data (feature engineering without a clear target variable defined in the transformation process), it is likely that this dataset is intended for unsupervised learning or the target was processed separately. The columns in `df_transformed` are the features after dimensionality reduction and feature selection. Therefore, I will treat all columns as features `X` and indicate that this is likely an unsupervised task.



````python
# Since no explicit target variable was defined or kept during the transformation,
# we assume this dataset is prepared for unsupervised learning or the target
# variable is handled separately.
# All columns in df_transformed are treated as features.
X = df_transformed.copy()

print("‚úÖ Features (X) prepared.")
print("\nNote: No explicit target variable (y) was identified in the transformed data.")
print("This dataset appears to be prepared for unsupervised learning or the target variable is handled separately.")

# For unsupervised learning, splitting into train/test sets is not typically required
# in the same way as supervised learning. However, if a split is needed for
# validation of unsupervised models or other purposes, it can be done on X.
# As the instructions mention splitting if a target is identified, and none was,
# we will skip the train/test split step as per the instructions.
````

## Choose and train model

### Subtask:
Select a suitable machine learning model and train it on the training data.


**Reasoning**:
Import the KMeans class and instantiate and fit a KMeans model to the feature data.



````python
from sklearn.cluster import KMeans

# Instantiate KMeans model with 5 clusters and a random state for reproducibility
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)

# Fit the model to the feature data
kmeans.fit(X)

print("‚úÖ KMeans model instantiated and fitted successfully!")
````

## Evaluate model

### Subtask:
Evaluate the trained model's performance on the testing data.


**Reasoning**:
Calculate and print the inertia and extract and print the first few cluster labels from the fitted KMeans model.



````python
# Calculate the inertia
inertia = kmeans.inertia_
print(f"Inertia (Within-Cluster Sum of Squares): {inertia}")

# Get the cluster labels
cluster_labels = kmeans.labels_

# Print the first few cluster labels
print("\nFirst 10 Cluster Labels:")
print(cluster_labels[:10])
````
